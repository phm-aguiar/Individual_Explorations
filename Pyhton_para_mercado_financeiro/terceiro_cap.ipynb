```python
import xlrd

wb = xlrd.open_workbook('EstatDad.xls')
plan = wb.sheet_by_name('Sheet1')

lin = plan.nrows
col = plan.ncols

dados = []

for i in range(col):
	coluna = plan.col_values(i)
	dados.append(coluna)
media = st.mean(dados[1][:])
mediana = st.median(dados[1][:])
desvioP = st.pstdev(dados[1][:])
desvioA = st.stdev(dados[1][:])

print(f"media = {media}\nmediana = {mediana}\ndesvio padrao populacional = {desvioP}\ndesvio padrao amostral = {desvioA}")
```

```python
plt.figure()
plt.hist(dados[1][:], bins = 3, color = 'gray')
plt.xlabel('classes')
plt.ylabel('frequencia')
```

```python
import random as rd

aleat = []

for i in range(100):
	aleat.append(rd.randint(1,60))

plt.plot(aleat)
```

```python
aleat = []
for i in range(100):
	aleat.append(rd.random())

plt.plot(aleat, color='black')
plt.xlabel('eixo horizontal')
plt.ylabel('nmeros aleatrios entre 0 e 1')
```

```python

aleat = []

for i in range(100):
	aleat.append(rd.uniform(-1,1))
 
plt.plot(aleat, color='black')
plt.xlabel('eixo horizontal')
plt.ylabel('nmeros aleatrios entre -1 e 1')
```

```python
aleat = []

for i in range(1000):
    aleat.append(rd.uniform(-1,1))

plt.subplot(211)
plt.plot(aleat, color='black')
plt.xlabel('eixo horizontal')
plt.ylabel('nmeros aleatrios entre -1 e 1')

plt.subplot(212)
plt.hist(aleat, bins=20, color='gray')
plt.xlabel('classes')
plt.ylabel('frequencia')
```
