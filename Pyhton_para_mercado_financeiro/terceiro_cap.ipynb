```python
from openpyxl import load_workbook

workbook = load_workbook('ibov.xlsx')

ibov = workbook['IBOV']
```

```python
fechamento = []
for linha in ibov.iter_rows(min_row=2, min_col=2, max_col=2):
	for celula in linha:
		fechamento.append(celula.value)
dias = []

for i in ibov.iter_rows(min_row=2, min_col=1, max_col=1):
	for j in i:
		dias.append(j.value)

print(dias)
```

```python
dados = ny.array(fechamento)

retorno = (dados[1:] - dados[:-1])/dados[0:-1] 

fig.subplot(211)
fig.plot(dias, dados, color='black')
fig.xlabel('dias')
fig.title('fechamento diario IBOV 2023 - 2024')

fig.subplot(212)
fig.plot(dias[1:], retorno, color='black')
fig.xlabel('dias')
fig.title('Retorno diario IBOV 2023 - 2024')
```

```python
fig.subplot(211)
fig.hist(retorno, bins=20, color='gray', density=True)
fig.xlabel('classes')
fig.ylabel('frequencia')
fig.title('Histograma do retorno diario IBOV 2023 - 2024')
xmin, xmax = fig.xlim()
media = st.mean(retorno)
desvio = st.pstdev(retorno)
eixo_x = ny.linspace(xmin, xmax, 100)
eixo_y = norm.pdf(eixo_x, media, desvio)
eixo_y2 = norm.cdf(eixo_x, media, desvio)
fig.plot(eixo_x, eixo_y, color='red')
# fig.plot(eixo_x, eixo_y2, color='blue')

import scipy.stats as stats
fig.subplot(212)
stats.probplot(retorno, dist='norm', plot=fig)
fig.title('QQ-plot do retorno diario IBOV 2023 - 2024')
```

```python
workbook = ox.load_workbook('ibov.xlsx')

ibov = workbook['IBOV']

fechamento = []
dias = []

for linha in ibov.iter_rows(min_row=2, min_col=2, max_col=2):
	for celula in linha:
		fechamento.append(celula.value)

for i in ibov.iter_rows(min_row=2, min_col=1, max_col=1):
	for j in i:
		dias.append(j.value)

workbook = ox.load_workbook('PETR4.xlsx')

petr4 = workbook['PETR4']

fechamento2 = []
dias2 = []

for linha in petr4.iter_rows(min_row=2, min_col=3, max_col=3):
	for celula in linha:
		fechamento2.append(celula.value)

for i in petr4.iter_rows(min_row=2, min_col=1, max_col=1):
	for j in i:
		dias2.append(j.value)
dias.pop()
```

```python
dados1 = ny.array(fechamento)
dados2 = ny.array(fechamento2)


beta, beta0, r_value, p_value, std_err = stats.linregress(dados1, dados2)
```

```python
ylin = beta*dados1 + beta0

fig.plot(dados1, ylin, '-k', dados1,dados2, 'ok')
fig.xlabel('IBOV')
fig.ylabel('PETR4')
```

```python
wb1 = ox.load_workbook('petrobras.xlsx')
wb2 = ox.load_workbook('vale.xlsx')

petr4 = wb1['PETR4.SA']
vale = wb2['VALE3']

fechamento1 = []
dias1 = []
fechamento2 = []
dias2 = []

for linha in petr4.iter_rows(min_row=2, min_col=2, max_col=2):
	for celula in linha:
		fechamento1.append(celula.value)

for i in petr4.iter_rows(min_row=2, min_col=1, max_col=1):
	for j in i:
		dias1.append(j.value)

for linha in vale.iter_rows(min_row=2, min_col=2, max_col=2):
	for celula in linha:
		fechamento2.append(celula.value)
for i in vale.iter_rows(min_row=2, min_col=1, max_col=1):
	for j in i:
		dias2.append(j.value)

dados1 = ny.array(fechamento1)
dados2 = ny.array(fechamento2)

beta, beta0, r_value, p_value, std_err = stats.linregress(dados1, dados2)

print(f"beta = {beta}\nbeta0 = {beta0}\nr_value = {r_value}\np_value = {p_value}\nstd_err = {std_err}")

ylin = beta*dados1 + beta0

fig.plot(dados1, ylin, '-k', dados1, dados2, 'ok')
fig.xlabel('PETR4')
fig.ylabel('VALE3')
```

```python
dad1 = [10, 11, 10, 9, 8, 9, 10, 11, 12, 9, 10, 8]

dad2 = [5, 6, 9, 10, 8, 5, 4, 5, 6, 7, 8, 5]
```

```python
stats.ttest_ind(dad1, dad2)
```

```python
fig.subplot(211)
fig.plot(dad1, '-k')
fig.plot(dad2, '--k')
fig.title('Comparao entre dad1 e dad2')

ax=fig.subplot(212)
ax.boxplot([dad1, dad2], showfliers=False)
ax.set_xticklabels(['dad1', 'dad2'])

print(stats.ttest_ind(dad1, dad2))
```

```python
retorno1 = (dados1[1:] - dados1[:-1])/dados1[:-1]
retorno2 = (dados2[1:] - dados2[:-1])/dados2[:-1]

print(stats.ttest_ind(retorno1, retorno2))
fig.subplot(311)
fig.plot(dias1[1:], retorno1, '-k')
fig.title('Retorno diario PETR4 2023 - 2024')
fig.subplot(312)
fig.plot(dias2[1:], retorno2, '-k')
fig.title('Retorno diario VALE3 2023 - 2024')

ax = fig.subplot(313)
ax.boxplot([retorno1, retorno2], showfliers=False)
ax.set_xticklabels(['PETR4', 'VALE3'])
```
